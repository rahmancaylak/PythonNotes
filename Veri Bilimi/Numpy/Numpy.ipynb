## NumPy Array'de İşlemler

### 1. Reshaping - Yeniden Şekillendirme
* Tek boyutlu arrrayi iki boyutluya, iki boyutlu arrayi de tek boyuta reshaping ile dönüştürebiliriz.

```python
np.arange(1,10)
```

```python
np.arange(1,10).reshape((3,3))
```

```python
# Bu yaptıklarımızı şimdi değişkenlere dökelim.
a = np.arange(1,10)
a
```

```python
a.ndim # a değişkeninin tek boyutlu olduğunu gördük
```

```python
b =  a.reshape(3,3) # reshape ile de b değişkenine boyutunu değiştirerek atadık.
b # Ve b değişkenini konsola yazdırdığımızda matrix görünümüne döndüğünü görebiliyoruz.
```

```python
b.ndim # b değişkeninin iki boyutlu olduğunu da görmüş olduk.
```

### 2. Concatenation - Array Birleştirme
* Elimizdeki arrayleri tek bir arrayde birleştirebiliriz.

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
# İki arrayı bu şekilde birleştirmiş olduk.
np.concatenate([x, y])
```

```python
# Birleştirdiğimiz arrayler tek boyutluydu, iki boyutluları da birleştirelim.
a = np.array([[1,2,3], 
              [4,5,6]])
np.concatenate([a,a])
```

```python
# Üstte satır olarak birleştirmiştik, axis sayesinde sütun olarakta birleştirebiliyoruz.
np.concatenate([a,a], axis = 1) # Axis yörüngeyi belirtir., 0 satır olarak, 1 ise sütun olarak birleştirir.
```

### 3. Splitting - Array Ayırma
* Belirttiğimiz değerlere göre arrayleri farklı değişkenlere parçalayabiliriz.

```python
x = np.array([1,2,3,99,99,3,2,1])
np.split(x, [3,5]) # 0'dan 2. index'e kadar, 3. indexten 4.index'e kadar ayırmasını söyledik.
# Yani burada 3 ve 5 dahil olmadığı için kafalar karışmasın aman dikkat. O yüzden 0'dan 2. index'e kadar değerler aldık.
```

```python
# Arrayleri parçaladık peki ama demiştik ki farklı değişkenlere parlaçayabiliriz. Aşağıya dikkat o zaman. :)
a,b,c = np.split(x, [3,5]) # üstte ayırma işlemlerini değişkenlere atadık.
```

## Slicing ile Elemanlara Erişmek (Array Alt Kümesine Erişmek)
* __Hatırlatma:__ değişken[x:y] ---> x: Başlangıç indexi, y: Bitiş indexi
* Bu konuya alt elemanlarda işlemler yapabilmek için öğreniyoruz ve hatırlıyoruz. Bir sonraki konu bunun üzerinde olacak.

```python
a = np.arange(20,30) #20'den 0'a kadar değerler oluşturduk.
a
```

```python
a[0:3] # 0. index'den 3.index'e kadar değerleri al dedik ama 3. index dahil değil.
```

```python
a[:3] # Eğer başlangıç değeri belirtilmezse, 0.index'den başlar.
```

```python
a[3:] # Eğer son değer belirtilmezse, başlangıç değerinden sonraki tüm indexleri alır.
```

```python
a[1::2] # 1.index'den başla ve 2'şer 2'şer artarak indexleri al. Yani 1, 3, 5, 7...
```

```python
a[0::2] # Üsttekine benzer ama 0.index'den başlatıyoruz. Böylece indexler 0, 2, 4, 6...
```

Şimdi iki boyutlu arraylerde slicing ile elamanlara erişelim.

```python
m = np.random.randint(10, size = (5,5))
m
```

```python
m[:,0] # 1. sütunun tamamını almış olduk.
```

```python
m[:,1] # 2.sütunun tamamını aldık. Yani, buradan anlayacağımız bitiş indexine yazdığımız değerler sütunları temsil ediyor.
```

```python
m[0,:] # 1. satırın tamamını aldık. Buradan anlayacağımız da başlangıç değeri satırları temsil ediyor.
```

```python
m[0:2, 0:3] # ilk iki satırı ve ilk 3 sütundaki elamanları al demiş olduk.
```

## Alt Küme Üzerinde İşlem Yapmak

```python
# Şimdi yukarıdaki işlemi uyguladığımızda ana arrayin değerlerini de değiştiriyor.
a
```

## Fancy Index ile Elemanlara Erişmek

```python
v = np.arange(0, 30, 3)
v
```

```python
# Index ile değerlere basit bir şekilde ulaşıyorduk.
v[3]
```

```python
# Farklı bir şey yaparak indexleri gruplayabiliriz. Gruplanan indexlerin de değerlerini geri döndürebiliriz.
[v[1], v[3], v[5]]
```

```python
getValues = [1,4,3]
v
```

```python
# Aşağıda yaptığımız şey, belirttiğimiz array'in indexine getValues'den gelen değerleri ata.
v[getValues]
# v[1], v[4], v[3] şeklinde olmuş oldu. Bu indexlere karşılık gelen değerlere de bu şekilde ulaşabildik.
# Bu işleme fancy index denir.
```

```python
m = np.arange(9).reshape((3,3))
m
```

```python
# 2 boyutlu fancy index için satır ve sütunlarımızı belirttik.
satir = np.array([0,1])
sutun = np.array([1,2])
```

```python
m[satir, sutun]
# m[[0,1],[1,2]] bu işlemi yapmış olduk.
```

```python
m
```

```python
# Basit index ile fancy kullandık.
m[0, [1,2]]
```

```python
# Slice ile de fancy kullanabiliriz.
m[0:, [1,2]]
```
